```python
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, QuantumRegister, ClassicalRegister, execute
from qiskit.visualization import *
from qiskit.quantum_info import random_statevector, Statevector
import numpy as np
```

```python
# Built a Quantum Circuit with 2 Qubit lines few Hadamard and CNOt gates
qr = QuantumRegister(2, 'qreg')
qc_circ_simple = QuantumCircuit(qr)
qc_circ_simple.h([0,1])
qc_circ_simple.x(0)
qc_circ_simple.z(1)
qc_circ_simple.draw()
```

```python
# choose  Are.statevector_simulator as bkend 
bkend_statevector = Aer.get_backend('statevector_simulator') 
result = execute(qc_circ_simple, bkend_statevector).result() 
statevector_simple = result.get_statevector(0, 2) # gets unitary matrix with Result from Job done above
print(statevector_simple)
plot_bloch_multivector(qc_circ_simple)
```

```python
plot_state_qsphere(qc_circ_simple)
```

```python
# let's do this excersize with an entangled qubit state
# Built a Quantum Circuit with 2 Qubit lines few Hadamard and CNOt gates
qr = QuantumRegister(2, 'qreg')
qc_circ = QuantumCircuit(qr)
qc_circ.h(0)
qc_circ.cx(0, 1)
qc_circ.draw()
```

```python
# assign statevector simulator as bkend 
bkend_statevector = Aer.get_backend('statevector_simulator') 
result = execute(qc_circ, bkend_statevector).result() 
statevector = result.get_statevector() # gets unitary matrix with Result from Job done above
print(statevector)
```

```python
# Defining Statevector Simulator as bkend needed for Getting statevector
sv_backend = Aer.get_backend('statevector_simulator') 
# Executes job to get the Statevector at the end of this circuit
sv_circ = execute(qc_circ, sv_backend).result().get_statevector(0,2) 
print(sv_circ) 
plot_state_qsphere(sv_circ)
```
